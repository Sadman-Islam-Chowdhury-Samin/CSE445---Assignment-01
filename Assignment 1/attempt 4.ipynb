```python


def compute_cost(X, y, w):
    m = X.shape[0]
    predictions = np.dot(X, w)
    cost = np.sum((predictions - y) ** 2)
    return cost / (2 * m)

def compute_gradient(X, y, w):
    m = X.shape[0]
    X = np.array(X, dtype=np.float64)  # Ensure X is a NumPy array of type float64
    w = np.array(w, dtype=np.float64)  # Ensure w is a NumPy array of type float64
    predictions = np.dot(X, w)
    gradient = np.dot(X.T, (predictions - y)) / m
    return gradient

def gradient_descent(X, y, w_init, alpha, num_iters, cost_function, gradient_function):
    w = copy.deepcopy(w_init)
    cost_history = []
    
    for i in range(num_iters):
        gradient = gradient_function(X, y, w)
        w -= alpha * gradient
        
        cost = cost_function(X, y, w)
        cost_history.append(cost)
        
        if i % math.ceil(num_iters / 10) == 0:
            print(f"Iteration {i:4}: Cost {cost:0.2e}, Parameters: {w}")
    
    return w, cost_history

# Load your data
#data = pd.read_csv('your_dataset.csv')  # Replace 'your_dataset.csv' with your dataset filename

# Remove rows with non-numeric values
data_numeric = data.apply(pd.to_numeric, errors='coerce')
data_numeric = data_numeric.dropna()

# Separate features (X) and target (y) from the numeric data
X = data_numeric.iloc[:, :-3].values
y = data_numeric.iloc[:, -1].values

# Convert X, y, and w_init to NumPy arrays
X = np.array(X, dtype=np.float64)
y = np.array(y, dtype=np.float64)

# Initialize parameters
num_features = X.shape[1]
w_init = np.zeros(num_features, dtype=np.float64)  # Initialize weights to zeros or any other value

# Set hyperparameters
alpha = 0.0001
num_iters = 10000

# Run gradient descent
w_final, cost_history = gradient_descent(X, y, w_init, alpha, num_iters, compute_cost, compute_gradient)

# Calculate Mean Squared Error (MSE)
mse = compute_cost(X, y, w_final)
print("Final Mean Squared Error:", mse)
```
